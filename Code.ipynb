# Handel missing value

```python
cat_cols = df.select_dtypes(include = ["object"]).columns
num_cols = df.select_dtypes(include = ["number"]).columns
```

#### SimpleImputer = Replace missing values using a descriptive statistic (e.g. mean, median, or most frequent) along each column, or using a constant value

```python
from sklearn.impute import SimpleImputer
num_impute = SimpleImputer(strategy="mean")
df[num_cols] = num_impute.fit_transform(df[num_cols])

col_impute = SimpleImputer(strategy="most_frequent")
df[cat_cols] = col_impute.fit_transform(df[cat_cols])
```
